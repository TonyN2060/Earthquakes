```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample rate (Hz) and duration (seconds)
sample_rate = 1000  # 1000 samples per second
duration = 10       # 10 seconds

# Time vector
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# Create a synthetic seismic waveform as a sum of sinusoids
frequency1 = 5  # Frequency in Hz
frequency2 = 10
amplitude1 = 1.0
amplitude2 = 0.5
seismic_waveform = amplitude1 * np.sin(2 * np.pi * frequency1 * t) + amplitude2 * np.sin(2 * np.pi * frequency2 * t)
```

```python
# Perform Fast Fourier Transform (FFT)
fft_result = np.fft.fft(seismic_waveform)

# Frequency values corresponding to the FFT result
frequencies = np.fft.fftfreq(len(fft_result), 1 / sample_rate)

# Calculate the magnitude (amplitude) of the FFT result
magnitude = np.abs(fft_result)
```

```python
# Plot the seismic waveform
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.plot(t, seismic_waveform)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Seismic Waveform')

# Plot the frequency spectrum
plt.subplot(122)
plt.plot(frequencies, magnitude)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Frequency Spectrum')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic seismic data with seismic waves and noise
np.random.seed(0)
t = np.linspace(0, 10, 1000)  # Time vector (0 to 10 seconds)
seismic_waves = 0.5 * np.sin(2 * np.pi * 1.0 * t) + 0.2 * np.sin(2 * np.pi * 3.0 * t)
noise = 0.1 * np.random.randn(1000)  # Gaussian noise
seismic_data = seismic_waves + noise

# Perform Fourier transform to analyze frequency content
freq_spectrum = np.fft.fft(seismic_data)
frequencies = np.fft.fftfreq(len(seismic_data), 1.0 / 1000)  # Sampling rate is 1000 Hz

# Visualize the original data and its frequency spectrum
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, seismic_data, label='Seismic Data (with Noise)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Seismic Data with Noise')

plt.subplot(2, 1, 2)
plt.plot(frequencies, np.abs(freq_spectrum))

plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Frequency Spectrum')

plt.tight_layout()
plt.show()
```
